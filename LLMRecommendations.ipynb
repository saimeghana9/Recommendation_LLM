```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta
import json
from google.colab import drive
from google.colab import files
import os

# Mount Google Drive
drive.mount('/content/drive')

# Set random seed for reproducibility
np.random.seed(42)
random.seed(42)

# =============================================================================
# 1. MOVIES DATASET
# =============================================================================

def generate_movies_data(n=5000):
    """Generate movies dataset with cross-domain recommendation capabilities"""

    # Movie data templates
    genres = ['Action', 'Comedy', 'Drama', 'Horror', 'Romance', 'Thriller', 'Sci-Fi',
              'Fantasy', 'Adventure', 'Crime', 'Mystery', 'Animation', 'Family',
              'Documentary', 'Biography', 'History', 'War', 'Western', 'Musical', 'Sport']

    moods = ['uplifting', 'melancholic', 'energetic', 'calm', 'romantic', 'mysterious',
             'inspiring', 'dark', 'funny', 'dramatic', 'peaceful', 'intense']

    weather_conditions = ['sunny', 'rainy', 'snowy', 'cloudy', 'stormy', 'foggy']

    time_of_day = ['morning', 'afternoon', 'evening', 'night', 'late_night']

    directors = ['Christopher Nolan', 'Steven Spielberg', 'Martin Scorsese', 'Quentin Tarantino',
                'Ridley Scott', 'James Cameron', 'Peter Jackson', 'David Fincher', 'Tim Burton',
                'Wes Anderson', 'Denis Villeneuve', 'Damien Chazelle', 'Greta Gerwig',
                'Jordan Peele', 'Ava DuVernay', 'Ryan Coogler', 'Taika Waititi', 'Bong Joon-ho']

    countries = ['USA', 'UK', 'France', 'Germany', 'Japan', 'South Korea', 'India', 'Italy',
                'Spain', 'Canada', 'Australia', 'Brazil', 'Mexico', 'Russia', 'China']

    languages = ['English', 'Spanish', 'French', 'German', 'Japanese', 'Korean', 'Hindi',
                'Italian', 'Portuguese', 'Russian', 'Chinese', 'Arabic']

    movies = []

    for i in range(n):
        # Basic info
        genre = random.choice(genres)
        year = random.randint(1940, 2025)
        duration = random.randint(80, 180)

        # Generate title based on genre
        if genre == 'Action':
            title = f"{random.choice(['The', 'Operation', 'Mission', 'Battle'])} {random.choice(['Thunder', 'Storm', 'Fire', 'Shadow'])}"
        elif genre == 'Comedy':
            title = f"{random.choice(['Funny', 'Crazy', 'Wild', 'Silly'])} {random.choice(['Adventures', 'Stories', 'Tales', 'Mishaps'])}"
        elif genre == 'Romance':
            title = f"{random.choice(['Love', 'Heart', 'Soul', 'Forever'])} {random.choice(['Story', 'Journey', 'Dream', 'Promise'])}"
        elif genre == 'Horror':
            title = f"{random.choice(['The', 'Night', 'Dark', 'Silent'])} {random.choice(['Scream', 'Whisper', 'Shadow', 'Curse'])}"
        else:
            title = f"{random.choice(['The', 'A', 'In', 'Beyond'])} {random.choice(['Journey', 'Story', 'Dream', 'Quest'])}"

        # Cross-domain attributes
        mood = random.choice(moods)
        weather_suitable = random.choice(weather_conditions)
        time_suitable = random.choice(time_of_day)

        # Rating and popularity
        rating = round(random.uniform(3.0, 9.5), 1)
        votes = random.randint(100, 1000000)

        # Description with cross-domain keywords
        description = f"A {mood} {genre.lower()} film that's perfect for {weather_suitable} {time_suitable}. " \
                     f"Directed by {random.choice(directors)}, this movie explores themes of {random.choice(['love', 'redemption', 'adventure', 'mystery', 'friendship'])}."

        movie = {
            'movie_id': f"MOV_{i+1:05d}",
            'title': title,
            'release_year': year,
            'genre': genre,
            'rating': rating,
            'votes': votes,
            'duration': duration,
            'description': description,
            'director': random.choice(directors),
            'cast': f"{random.choice(['Tom Hanks', 'Meryl Streep', 'Leonardo DiCaprio', 'Emma Stone'])} and {random.choice(['Brad Pitt', 'Jennifer Lawrence', 'Ryan Gosling', 'Scarlett Johansson'])}",
            'country': random.choice(countries),
            'language': random.choice(languages),
            'certificate': random.choice(['G', 'PG', 'PG-13', 'R', 'NC-17']),
            'budget': random.randint(1000000, 300000000),
            'revenue': random.randint(1000000, 1000000000),
            'keywords': f"{mood},{weather_suitable},{time_suitable},{genre.lower()}",
            'mood': mood,
            'weather_suitable': weather_suitable,
            'time_suitable': time_suitable,
            'imdb_id': f"tt{random.randint(1000000, 9999999)}",
            'poster_url': f"https://example.com/posters/movie_{i+1}.jpg"
        }
        movies.append(movie)

    return pd.DataFrame(movies)

# =============================================================================
# 2. BOOKS DATASET
# =============================================================================

def generate_books_data(n=5000):
    """Generate books dataset with cross-domain recommendation capabilities"""

    genres = ['Fiction', 'Non-Fiction', 'Mystery', 'Romance', 'Sci-Fi', 'Fantasy',
              'Thriller', 'Biography', 'History', 'Self-Help', 'Philosophy', 'Poetry',
              'Adventure', 'Horror', 'Comedy', 'Drama', 'Educational', 'Cooking',
              'Travel', 'Art', 'Science', 'Business']

    moods = ['inspiring', 'thoughtful', 'relaxing', 'exciting', 'educational', 'romantic',
             'mysterious', 'uplifting', 'melancholic', 'humorous', 'dramatic', 'peaceful']

    reading_conditions = ['rainy_day', 'sunny_day', 'cozy_night', 'travel', 'study_time',
                         'bedtime', 'morning_read', 'beach_read', 'winter_read']

    authors = ['J.K. Rowling', 'Stephen King', 'Agatha Christie', 'Ernest Hemingway',
              'Jane Austen', 'Mark Twain', 'Virginia Woolf', 'Toni Morrison', 'Gabriel Garcia Marquez',
              'Haruki Murakami', 'Chimamanda Ngozi Adichie', 'Zadie Smith', 'Kazuo Ishiguro',
              'Margaret Atwood', 'Neil Gaiman', 'George R.R. Martin', 'J.R.R. Tolkien']

    publishers = ['Penguin Random House', 'HarperCollins', 'Simon & Schuster', 'Macmillan',
                 'Hachette', 'Scholastic', 'Oxford University Press', 'Cambridge University Press']

    books = []

    for i in range(n):
        genre = random.choice(genres)
        year = random.randint(1950, 2025)
        pages = random.randint(100, 800)

        # Generate title based on genre
        if genre == 'Mystery':
            title = f"The {random.choice(['Secret', 'Hidden', 'Lost', 'Missing'])} {random.choice(['Case', 'Affair', 'Mystery', 'Puzzle'])}"
        elif genre == 'Romance':
            title = f"{random.choice(['Love', 'Heart', 'Soul', 'Forever'])} {random.choice(['Story', 'Journey', 'Promise', 'Dream'])}"
        elif genre == 'Sci-Fi':
            title = f"{random.choice(['The', 'Beyond', 'Into', 'Through'])} {random.choice(['Stars', 'Void', 'Future', 'Unknown'])}"
        elif genre == 'Self-Help':
            title = f"How to {random.choice(['Achieve', 'Master', 'Discover', 'Transform'])} {random.choice(['Success', 'Happiness', 'Peace', 'Growth'])}"
        else:
            title = f"{random.choice(['The', 'A', 'In', 'Beyond'])} {random.choice(['Journey', 'Story', 'Quest', 'Adventure'])}"

        # Cross-domain attributes
        mood = random.choice(moods)
        reading_condition = random.choice(reading_conditions)

        # Rating and popularity
        rating = round(random.uniform(3.0, 5.0), 1)
        ratings_count = random.randint(10, 100000)

        # Description with cross-domain keywords
        description = f"An {mood} {genre.lower()} book perfect for {reading_condition}. " \
                     f"This {mood} story explores themes of {random.choice(['love', 'adventure', 'mystery', 'growth', 'friendship', 'discovery'])}."

        book = {
            'book_id': f"BOOK_{i+1:05d}",
            'title': title,
            'author': random.choice(authors),
            'genre': genre,
            'published_year': year,
            'pages': pages,
            'description': description,
            'publisher': random.choice(publishers),
            'language': 'English',
            'average_rating': rating,
            'ratings_count': ratings_count,
            'isbn13': f"978{random.randint(1000000000000, 9999999999999)}",
            'isbn10': f"{random.randint(1000000000, 9999999999)}",
            'thumbnail_url': f"https://example.com/covers/book_{i+1}.jpg",
            'downloads': random.randint(100, 1000000),
            'mood': mood,
            'reading_condition': reading_condition,
            'keywords': f"{mood},{reading_condition},{genre.lower()}"
        }
        books.append(book)

    return pd.DataFrame(books)

# =============================================================================
# 3. FOOD/RECIPES DATASET
# =============================================================================

def generate_food_data(n=5000):
    """Generate food/recipes dataset with cross-domain recommendation capabilities"""

    categories = ['Appetizer', 'Main Course', 'Dessert', 'Soup', 'Salad', 'Beverage',
                 'Breakfast', 'Lunch', 'Dinner', 'Snack', 'Side Dish', 'Sauce']

    cuisines = ['Italian', 'Chinese', 'Indian', 'Mexican', 'French', 'Japanese', 'Thai',
               'American', 'Mediterranean', 'Korean', 'Vietnamese', 'Spanish', 'Greek',
               'Lebanese', 'Turkish', 'Moroccan', 'Brazilian', 'Peruvian']

    moods = ['comforting', 'energizing', 'relaxing', 'celebratory', 'healthy', 'indulgent',
             'light', 'hearty', 'refreshing', 'warming', 'cooling', 'satisfying']

    occasions = ['rainy_day', 'sunny_day', 'date_night', 'family_dinner', 'party', 'workout',
                'study_session', 'movie_night', 'picnic', 'winter_day', 'summer_day', 'hangover']

    ingredients_base = ['chicken', 'beef', 'fish', 'vegetables', 'rice', 'pasta', 'eggs',
                       'cheese', 'tomatoes', 'onions', 'garlic', 'herbs', 'spices', 'bread',
                       'potatoes', 'beans', 'lentils', 'mushrooms', 'spinach', 'carrots']

    food_items = []

    for i in range(n):
        category = random.choice(categories)
        cuisine = random.choice(cuisines)

        # Generate name based on category and cuisine
        if category == 'Main Course':
            name = f"{cuisine} {random.choice(['Chicken', 'Beef', 'Fish', 'Vegetable'])} {random.choice(['Stir-fry', 'Curry', 'Stew', 'Grill'])}"
        elif category == 'Dessert':
            name = f"{cuisine} {random.choice(['Chocolate', 'Vanilla', 'Berry', 'Citrus'])} {random.choice(['Cake', 'Pudding', 'Tart', 'Mousse'])}"
        elif category == 'Soup':
            name = f"{cuisine} {random.choice(['Tomato', 'Vegetable', 'Chicken', 'Mushroom'])} Soup"
        else:
            name = f"{cuisine} {category}"

        # Cross-domain attributes
        mood = random.choice(moods)
        occasion = random.choice(occasions)

        # Cooking details
        prep_time = random.randint(10, 120)
        cook_time = random.randint(15, 180)
        servings = random.randint(1, 12)
        difficulty = random.choice(['Easy', 'Medium', 'Hard'])

        # Nutritional info
        calories = random.randint(100, 800)
        rating = round(random.uniform(3.0, 5.0), 1)
        review_count = random.randint(5, 5000)

        # Ingredients
        num_ingredients = random.randint(3, 15)
        ingredients = random.sample(ingredients_base, min(num_ingredients, len(ingredients_base)))

        # Description with cross-domain keywords
        description = f"A {mood} {cuisine.lower()} {category.lower()} perfect for {occasion}. " \
                     f"This {mood} dish is {random.choice(['delicious', 'nutritious', 'flavorful', 'satisfying'])} and great for {occasion}."

        food_item = {
            'recipe_id': f"FOOD_{i+1:05d}",
            'name': name,
            'author': f"Chef {random.choice(['Maria', 'John', 'Sarah', 'David', 'Lisa', 'Michael'])}",
            'description': description,
            'category': category,
            'cuisine_type': cuisine,
            'keywords': f"{mood},{occasion},{cuisine.lower()},{category.lower()}",
            'ingredients': ', '.join(ingredients),
            'cooking_time': cook_time,
            'prep_time': prep_time,
            'total_time': prep_time + cook_time,
            'servings': servings,
            'calories': calories,
            'rating': rating,
            'review_count': review_count,
            'difficulty_level': difficulty,
            'mood': mood,
            'occasion': occasion,
            'instructions': f"1. Prepare ingredients. 2. Cook according to {cuisine} tradition. 3. Serve hot and enjoy!",
            'image_urls': f"https://example.com/food/recipe_{i+1}.jpg"
        }
        food_items.append(food_item)

    return pd.DataFrame(food_items)

# =============================================================================
# 4. MUSIC DATASET
# =============================================================================

def generate_music_data(n=5000):
    """Generate music dataset with cross-domain recommendation capabilities"""

    genres = ['Pop', 'Rock', 'Hip-Hop', 'Jazz', 'Classical', 'Electronic', 'Country',
              'R&B', 'Blues', 'Folk', 'Reggae', 'Punk', 'Metal', 'Indie', 'Alternative',
              'Ambient', 'Funk', 'Soul', 'Gospel', 'World', 'Latin', 'K-Pop']

    moods = ['happy', 'sad', 'energetic', 'calm', 'romantic', 'melancholic', 'uplifting',
             'aggressive', 'peaceful', 'nostalgic', 'mysterious', 'dramatic', 'playful',
             'contemplative', 'euphoric', 'angry', 'hopeful', 'dreamy', 'intense']

    activities = ['workout', 'study', 'cooking', 'driving', 'sleeping', 'party', 'meditation',
                 'cleaning', 'reading', 'gaming', 'walking', 'showering', 'working', 'relaxing']

    weather_moods = ['rainy_day', 'sunny_day', 'snowy_day', 'stormy_night', 'foggy_morning']

    artists = ['Taylor Swift', 'Ed Sheeran', 'Billie Eilish', 'Drake', 'Ariana Grande',
              'The Weeknd', 'Post Malone', 'Dua Lipa', 'Harry Styles', 'Olivia Rodrigo',
              'Bad Bunny', 'SZA', 'Kendrick Lamar', 'Lana Del Rey', 'Bruno Mars',
              'Adele', 'Coldplay', 'Imagine Dragons', 'Maroon 5', 'OneRepublic']

    music_tracks = []

    for i in range(n):
        genre = random.choice(genres)
        year = random.randint(1940, 2025)

        # Generate title based on genre and mood
        mood = random.choice(moods)
        if mood == 'happy':
            title = f"{random.choice(['Sunshine', 'Joy', 'Smile', 'Bright'])} {random.choice(['Days', 'Moments', 'Times', 'Life'])}"
        elif mood == 'sad':
            title = f"{random.choice(['Tears', 'Rain', 'Goodbye', 'Lonely'])} {random.choice(['Tonight', 'Forever', 'Heart', 'Soul'])}"
        elif mood == 'energetic':
            title = f"{random.choice(['Fire', 'Thunder', 'Power', 'Energy'])} {random.choice(['Storm', 'Rush', 'Beat', 'Pulse'])}"
        else:
            title = f"{random.choice(['The', 'In', 'Through', 'Beyond'])} {random.choice(['Dream', 'Night', 'Day', 'Moment'])}"

        # Cross-domain attributes
        activity = random.choice(activities)
        weather_mood = random.choice(weather_moods)

        # Audio features
        tempo = random.randint(60, 180)
        energy = round(random.uniform(0.0, 1.0), 3)
        danceability = round(random.uniform(0.0, 1.0), 3)
        valence = round(random.uniform(0.0, 1.0), 3)
        acousticness = round(random.uniform(0.0, 1.0), 3)
        instrumentalness = round(random.uniform(0.0, 1.0), 3)
        loudness = round(random.uniform(-20.0, 0.0), 1)
        speechiness = round(random.uniform(0.0, 1.0), 3)
        liveness = round(random.uniform(0.0, 1.0), 3)

        # Popularity and metadata
        popularity = random.randint(0, 100)
        duration = random.randint(120, 360)  # 2-6 minutes

        # Generate lyrics snippet based on mood
        if mood == 'happy':
            lyrics = "Sunshine in my heart, everything's alright, dancing through the day and into the night"
        elif mood == 'sad':
            lyrics = "Tears falling like rain, memories remain, walking through the pain alone again"
        elif mood == 'romantic':
            lyrics = "You're the light in my darkness, the love in my heart, forever we'll be together, never apart"
        else:
            lyrics = "Through the journey of life, we find our way, in every moment, in every day"

        music_track = {
            'track_id': f"MUSIC_{i+1:05d}",
            'title': title,
            'artist': random.choice(artists),
            'genre': genre,
            'release_year': year,
            'album': f"{title} - Album",
            'lyrics': lyrics,
            'tempo': tempo,
            'energy': energy,
            'danceability': danceability,
            'valence': valence,
            'acousticness': acousticness,
            'instrumentalness': instrumentalness,
            'loudness': loudness,
            'speechiness': speechiness,
            'liveness': liveness,
            'popularity': popularity,
            'key': random.randint(0, 11),
            'mode': random.choice([0, 1]),
            'time_signature': random.choice([3, 4, 5, 6, 7]),
            'duration': duration,
            'language': random.choice(['English', 'Spanish', 'French', 'Korean', 'Japanese']),
            'mood': mood,
            'activity': activity,
            'weather_mood': weather_mood,
            'keywords': f"{mood},{activity},{weather_mood},{genre.lower()}"
        }
        music_tracks.append(music_track)

    return pd.DataFrame(music_tracks)

# =============================================================================
# 5. TV SHOWS DATASET
# =============================================================================

def generate_tv_shows_data(n=5000):
    """Generate TV shows dataset with cross-domain recommendation capabilities"""

    genres = ['Drama', 'Comedy', 'Action', 'Thriller', 'Sci-Fi', 'Fantasy', 'Horror',
              'Romance', 'Crime', 'Mystery', 'Documentary', 'Reality', 'Animation',
              'Adventure', 'Biography', 'History', 'War', 'Western', 'Musical', 'Sport']

    show_types = ['TV Series', 'TV Mini Series', 'TV Movie', 'TV Special', 'Documentary Series']

    moods = ['uplifting', 'dark', 'funny', 'dramatic', 'mysterious', 'romantic', 'intense',
             'lighthearted', 'thoughtful', 'exciting', 'calm', 'melancholic', 'inspiring']

    viewing_conditions = ['rainy_day', 'sunny_day', 'cozy_night', 'family_time', 'date_night',
                         'binge_watching', 'background_noise', 'focused_watching', 'workout',
                         'study_break', 'bedtime', 'weekend_marathon']

    networks = ['Netflix', 'HBO', 'Disney+', 'Amazon Prime', 'Hulu', 'Apple TV+', 'Paramount+',
               'Peacock', 'Showtime', 'Starz', 'FX', 'AMC', 'BBC', 'CBS', 'NBC', 'ABC']

    countries = ['USA', 'UK', 'Canada', 'Australia', 'Germany', 'France', 'Japan', 'South Korea',
                'Spain', 'Italy', 'Brazil', 'Mexico', 'India', 'Sweden', 'Norway', 'Denmark']

    tv_shows = []

    for i in range(n):
        genre = random.choice(genres)
        show_type = random.choice(show_types)
        year = random.randint(1940, 2025)

        # Generate title based on genre
        if genre == 'Drama':
            title = f"{random.choice(['The', 'Breaking', 'Walking', 'Game of'])} {random.choice(['Story', 'Bad', 'Dead', 'Thrones'])}"
        elif genre == 'Comedy':
            title = f"{random.choice(['Friends', 'The Office', 'Parks and Recreation', 'Brooklyn Nine-Nine'])}"
        elif genre == 'Sci-Fi':
            title = f"{random.choice(['The', 'Star', 'Black', 'Dark'])} {random.choice(['Expanse', 'Trek', 'Mirror', 'Matter'])}"
        elif genre == 'Crime':
            title = f"{random.choice(['True', 'The', 'Breaking', 'Better Call'])} {random.choice(['Detective', 'Wire', 'Bad', 'Saul'])}"
        else:
            title = f"{random.choice(['The', 'A', 'In', 'Beyond'])} {random.choice(['Series', 'Show', 'Chronicles', 'Tales'])}"

        # Cross-domain attributes
        mood = random.choice(moods)
        viewing_condition = random.choice(viewing_conditions)

        # Show details
        episodes = random.randint(1, 200)
        seasons = random.randint(1, 15)
        duration = random.randint(20, 60)  # minutes per episode
        rating = round(random.uniform(3.0, 9.5), 1)
        votes = random.randint(100, 1000000)

        # Description with cross-domain keywords
        description = f"A {mood} {genre.lower()} {show_type.lower()} perfect for {viewing_condition}. " \
                     f"This {mood} series explores themes of {random.choice(['love', 'adventure', 'mystery', 'friendship', 'redemption', 'discovery'])}."

        tv_show = {
            'show_id': f"TV_{i+1:05d}",
            'title': title,
            'type': show_type,
            'genre': genre,
            'release_year': year,
            'end_year': year + random.randint(1, 10) if random.random() > 0.3 else None,
            'episodes': episodes,
            'seasons': seasons,
            'duration': duration,
            'rating': rating,
            'votes': votes,
            'description': description,
            'director': f"Director {random.choice(['Smith', 'Johnson', 'Williams', 'Brown', 'Davis'])}",
            'cast': f"{random.choice(['Jennifer Aniston', 'Matthew Perry', 'Courteney Cox'])} and {random.choice(['David Schwimmer', 'Matt LeBlanc', 'Lisa Kudrow'])}",
            'network': random.choice(networks),
            'country': random.choice(countries),
            'language': random.choice(['English', 'Spanish', 'French', 'Korean', 'Japanese']),
            'certificate': random.choice(['TV-G', 'TV-PG', 'TV-14', 'TV-MA']),
            'status': random.choice(['Released', 'Ongoing', 'Cancelled', 'Ended']),
            'imdb_id': f"tt{random.randint(1000000, 9999999)}",
            'poster_url': f"https://example.com/posters/tv_{i+1}.jpg",
            'mood': mood,
            'viewing_condition': viewing_condition,
            'keywords': f"{mood},{viewing_condition},{genre.lower()}"
        }
        tv_shows.append(tv_show)

    return pd.DataFrame(tv_shows)

# =============================================================================
# MAIN EXECUTION
# =============================================================================

def main():
    print("üé¨ Generating Movies Dataset...")
    movies_df = generate_movies_data(5000)
    print(f"‚úÖ Generated {len(movies_df)} movie records")

    print("\nüìö Generating Books Dataset...")
    books_df = generate_books_data(5000)
    print(f"‚úÖ Generated {len(books_df)} book records")

    print("\nÔøΩÔøΩÔ∏è Generating Food/Recipes Dataset...")
    food_df = generate_food_data(5000)
    print(f"‚úÖ Generated {len(food_df)} food/recipe records")

    print("\nüéµ Generating Music Dataset...")
    music_df = generate_music_data(5000)
    print(f"‚úÖ Generated {len(music_df)} music records")

    print("\nüì∫ Generating TV Shows Dataset...")
    tv_shows_df = generate_tv_shows_data(5000)
    print(f"‚úÖ Generated {len(tv_shows_df)} TV show records")

    # Save to CSV files
    print("\nüíæ Saving datasets to CSV files...")

    # Create directory in Google Drive
    drive_path = '/content/drive/MyDrive/Recommendation_Datasets'
    os.makedirs(drive_path, exist_ok=True)

    # Save each dataset
    movies_df.to_csv(f'{drive_path}/movies.csv', index=False)
    books_df.to_csv(f'{drive_path}/books.csv', index=False)
    food_df.to_csv(f'{drive_path}/food.csv', index=False)
    music_df.to_csv(f'{drive_path}/music.csv', index=False)
    tv_shows_df.to_csv(f'{drive_path}/tv_shows.csv', index=False)

    print(f"‚úÖ All datasets saved to: {drive_path}")

    # Display sample data from each dataset
    print("\nüìä Sample Data Preview:")
    print("\nüé¨ Movies Sample:")
    print(movies_df[['title', 'genre', 'mood', 'weather_suitable', 'rating']].head(3))

    print("\nüìö Books Sample:")
    print(books_df[['title', 'genre', 'mood', 'reading_condition', 'average_rating']].head(3))

    print("\nÔøΩÔøΩÔ∏è Food Sample:")
    print(food_df[['name', 'cuisine_type', 'mood', 'occasion', 'rating']].head(3))

    print("\nüéµ Music Sample:")
    print(music_df[['title', 'genre', 'mood', 'activity', 'popularity']].head(3))

    print("\nüì∫ TV Shows Sample:")
    print(tv_shows_df[['title', 'genre', 'mood', 'viewing_condition', 'rating']].head(3))

    # Dataset statistics
    print(f"\nüìà Dataset Statistics:")
    print(f"Total Movies: {len(movies_df)}")
    print(f"Total Books: {len(books_df)}")
    print(f"Total Food/Recipes: {len(food_df)}")
    print(f"Total Music Tracks: {len(music_df)}")
    print(f"Total TV Shows: {len(tv_shows_df)}")
    print(f"Total Records: {len(movies_df) + len(books_df) + len(food_df) + len(music_df) + len(tv_shows_df)}")

    print(f"\nüéØ Cross-Domain Query Examples:")
    print("‚Ä¢ 'Movie for a rainy day' ‚Üí Filter movies with weather_suitable='rainy'")
    print("‚Ä¢ 'Music for sadness' ‚Üí Filter music with mood='sad'")
    print("‚Ä¢ 'Recipe for a good day' ‚Üí Filter food with mood='comforting' or 'energizing'")
    print("‚Ä¢ 'Music that goes with fried rice' ‚Üí Filter music with activity='cooking'")
    print("‚Ä¢ 'Book for bedtime' ‚Üí Filter books with reading_condition='bedtime'")
    print("‚Ä¢ 'TV show for family time' ‚Üí Filter TV shows with viewing_condition='family_time'")

    return movies_df, books_df, food_df, music_df, tv_shows_df

# Run the main function
if __name__ == "__main__":
    movies_df, books_df, food_df, music_df, tv_shows_df = main()
```

```python
# First, install required packages with correct versions
!pip install sentence-transformers lightgbm xgboost networkx transformers
!pip install langchain langchain-community faiss-cpu

import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer
import lightgbm as lgb
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import torch
import torch.nn as nn
from transformers import AutoModel, AutoTokenizer
import networkx as nx
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_community.llms import HuggingFacePipeline
from langchain.chains import RetrievalQA
import transformers
from typing import List, Dict, Any
import json

class CrossDomainRecommender:
    def __init__(self, movies_df, books_df, food_df, music_df, tv_shows_df):
        self.movies_df = movies_df
        self.books_df = books_df
        self.food_df = food_df
        self.music_df = music_df
        self.tv_shows_df = tv_shows_df

        # Initialize models
        self.tfidf_vectorizers = {}
        self.sentence_bert_model = SentenceTransformer('all-MiniLM-L6-v2')
        self.rag_embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

        # For multimodal alignment
        self.domain_embeddings = {}

        # For graph recommendations
        self.item_graph = nx.Graph()
        self.user_profiles = {}

    def prepare_domain_data(self):
        """Prepare data for each domain with combined text features"""
        # Movies
        self.movies_df['combined_text'] = (
            self.movies_df['title'] + ' ' +
            self.movies_df['description'] + ' ' +
            self.movies_df['genre'] + ' ' +
            self.movies_df['mood'] + ' ' +
            self.movies_df['keywords']
        )

        # Books
        self.books_df['combined_text'] = (
            self.books_df['title'] + ' ' +
            self.books_df['description'] + ' ' +
            self.books_df['genre'] + ' ' +
            self.books_df['mood'] + ' ' +
            self.books_df['keywords']
        )

        # Food
        self.food_df['combined_text'] = (
            self.food_df['name'] + ' ' +
            self.food_df['description'] + ' ' +
            self.food_df['cuisine_type'] + ' ' +
            self.food_df['mood'] + ' ' +
            self.food_df['keywords']
        )

        # Music
        self.music_df['combined_text'] = (
            self.music_df['title'] + ' ' +
            self.music_df['lyrics'] + ' ' +
            self.music_df['genre'] + ' ' +
            self.music_df['mood'] + ' ' +
            self.music_df['keywords']
        )

        # TV Shows
        self.tv_shows_df['combined_text'] = (
            self.tv_shows_df['title'] + ' ' +
            self.tv_shows_df['description'] + ' ' +
            self.tv_shows_df['genre'] + ' ' +
            self.tv_shows_df['mood'] + ' ' +
            self.tv_shows_df['keywords']
        )

    def train_tfidf_models(self):
        """Train TF-IDF models for each domain"""
        domains = {
            'movies': self.movies_df,
            'books': self.books_df,
            'food': self.food_df,
            'music': self.music_df,
            'tv_shows': self.tv_shows_df
        }

        for domain, df in domains.items():
            vectorizer = TfidfVectorizer(max_features=5000, stop_words='english')
            tfidf_matrix = vectorizer.fit_transform(df['combined_text'])
            self.tfidf_vectorizers[domain] = (vectorizer, tfidf_matrix)

    def content_based_recommendations(self, domain: str, query: str, n_recommendations: int = 5):
        """Content-based recommendations using TF-IDF and cosine similarity"""
        if domain not in self.tfidf_vectorizers:
            raise ValueError(f"Domain {domain} not found. Available domains: {list(self.tfidf_vectorizers.keys())}")

        vectorizer, tfidf_matrix = self.tfidf_vectorizers[domain]
        query_vec = vectorizer.transform([query])

        # Calculate cosine similarities
        similarities = cosine_similarity(query_vec, tfidf_matrix).flatten()

        # Get top N recommendations
        df = getattr(self, f"{domain}_df")
        top_indices = similarities.argsort()[-n_recommendations:][::-1]

        return df.iloc[top_indices]

    def sentence_bert_recommendations(self, domain: str, query: str, n_recommendations: int = 5):
        """Recommendations using Sentence-BERT embeddings"""
        df = getattr(self, f"{domain}_df")

        # Generate embeddings if not already done
        if domain not in self.domain_embeddings:
            texts = df['combined_text'].tolist()
            self.domain_embeddings[domain] = self.sentence_bert_model.encode(texts)

        # Encode query
        query_embedding = self.sentence_bert_model.encode([query])

        # Calculate similarities
        similarities = cosine_similarity(query_embedding, self.domain_embeddings[domain]).flatten()

        # Get top N recommendations
        top_indices = similarities.argsort()[-n_recommendations:][::-1]

        return df.iloc[top_indices]

    def train_ranking_model(self, domain: str, target_metric: str = 'rating'):
        """Train a LightGBM or XGBoost ranking model for a domain"""
        df = getattr(self, f"{domain}_df")

        # Prepare features - using a simplified approach
        # In practice, you'd want to engineer more meaningful features
        text_features = self.domain_embeddings[domain]

        # Create synthetic target for demonstration
        # In real scenarios, you'd have actual user interactions
        y = df[target_metric].values

        # Split data
        X_train, X_test, y_train, y_test = train_test_split(
            text_features, y, test_size=0.2, random_state=42
        )

        # Train LightGBM model
        model = lgb.LGBMRegressor(
            n_estimators=100,
            learning_rate=0.1,
            random_state=42
        )

        model.fit(X_train, y_train)

        return model, X_test, y_test

    def build_cross_domain_graph(self):
        """Build a graph connecting items across domains based on shared attributes"""
        # Add nodes for all items
        for idx, row in self.movies_df.iterrows():
            self.item_graph.add_node(f"movie_{row['movie_id']}",
                                   type='movie',
                                   attributes={'genre': row['genre'], 'mood': row['mood']})

        for idx, row in self.books_df.iterrows():
            self.item_graph.add_node(f"book_{row['book_id']}",
                                   type='book',
                                   attributes={'genre': row['genre'], 'mood': row['mood']})

        for idx, row in self.food_df.iterrows():
            self.item_graph.add_node(f"food_{row['recipe_id']}",
                                   type='food',
                                   attributes={'cuisine': row['cuisine_type'], 'mood': row['mood']})

        for idx, row in self.music_df.iterrows():
            self.item_graph.add_node(f"music_{row['track_id']}",
                                   type='music',
                                   attributes={'genre': row['genre'], 'mood': row['mood']})

        for idx, row in self.tv_shows_df.iterrows():
            self.item_graph.add_node(f"tv_{row['show_id']}",
                                   type='tv',
                                   attributes={'genre': row['genre'], 'mood': row['mood']})

        # Add edges based on shared attributes
        # This is a simplified version - in practice, you'd use more sophisticated similarity measures
        all_nodes = list(self.item_graph.nodes(data=True))

        for i, (node1, attrs1) in enumerate(all_nodes):
            for j, (node2, attrs2) in enumerate(all_nodes[i+1:], i+1):
                # Connect items with similar mood
                if (attrs1['attributes']['mood'] == attrs2['attributes']['mood'] and
                    node1 != node2):
                    self.item_graph.add_edge(node1, node2, weight=1.0)

                # Connect items with similar genre (if applicable)
                if ('genre' in attrs1['attributes'] and 'genre' in attrs2['attributes'] and
                    attrs1['attributes']['genre'] == attrs2['attributes']['genre'] and
                    node1 != node2):
                    self.item_graph.add_edge(node1, node2, weight=1.0)

    def graph_based_recommendations(self, seed_items: List[str], n_recommendations: int = 5):
        """Generate recommendations using graph traversal"""
        recommendations = {}

        for seed in seed_items:
            if seed in self.item_graph:
                # Get neighbors of the seed item
                neighbors = list(self.item_graph.neighbors(seed))

                # Score neighbors by degree and edge weights
                scores = {}
                for neighbor in neighbors:
                    if neighbor not in scores:
                        scores[neighbor] = 0

                    # Add edge weight
                    if self.item_graph.has_edge(seed, neighbor):
                        scores[neighbor] += self.item_graph[seed][neighbor].get('weight', 1.0)

                # Sort by score
                sorted_neighbors = sorted(scores.items(), key=lambda x: x[1], reverse=True)

                # Add to recommendations
                for neighbor, score in sorted_neighbors[:n_recommendations]:
                    if neighbor not in recommendations:
                        recommendations[neighbor] = 0
                    recommendations[neighbor] += score

        # Sort all recommendations
        sorted_recommendations = sorted(recommendations.items(), key=lambda x: x[1], reverse=True)

        return [rec[0] for rec in sorted_recommendations[:n_recommendations]]

    def setup_rag_system(self):
        """Set up the RAG system for conversational queries"""
        # Combine all domain data for the knowledge base
        all_data = []

        for idx, row in self.movies_df.iterrows():
            all_data.append(f"Movie: {row['title']}. Description: {row['description']}. Genre: {row['genre']}. Mood: {row['mood']}")

        for idx, row in self.books_df.iterrows():
            all_data.append(f"Book: {row['title']}. Description: {row['description']}. Genre: {row['genre']}. Mood: {row['mood']}")

        for idx, row in self.food_df.iterrows():
            all_data.append(f"Food: {row['name']}. Description: {row['description']}. Cuisine: {row['cuisine_type']}. Mood: {row['mood']}")

        for idx, row in self.music_df.iterrows():
            all_data.append(f"Music: {row['title']}. Lyrics: {row['lyrics'][:100]}. Genre: {row['genre']}. Mood: {row['mood']}")

        for idx, row in self.tv_shows_df.iterrows():
            all_data.append(f"TV Show: {row['title']}. Description: {row['description']}. Genre: {row['genre']}. Mood: {row['mood']}")

        # Create vector store
        self.vector_store = FAISS.from_texts(all_data, self.rag_embeddings)

        # Initialize LLM (simplified version without pipeline)
        # In a real implementation, you would set up a proper HuggingFace pipeline
        self.llm = None  # Placeholder - would need to set up properly

        # Create a simple retriever for demonstration
        self.retriever = self.vector_store.as_retriever(search_kwargs={"k": 3})

    def conversational_query(self, query: str):
        """Handle conversational queries using RAG (simplified version)"""
        # For demonstration, we'll just return the top matching documents
        docs = self.retriever.get_relevant_documents(query)

        # Format the response
        response = "Based on your query, here are some recommendations:\n\n"
        for i, doc in enumerate(docs):
            response += f"{i+1}. {doc.page_content[:200]}...\n\n"

        return response

    def multimodal_alignment(self):
        """Align embeddings across domains for cross-modal recommendations"""
        # Get embeddings for all domains
        domain_texts = {}
        domain_texts['movies'] = self.movies_df['combined_text'].tolist()
        domain_texts['books'] = self.books_df['combined_text'].tolist()
        domain_texts['food'] = self.food_df['combined_text'].tolist()
        domain_texts['music'] = self.music_df['combined_text'].tolist()
        domain_texts['tv_shows'] = self.tv_shows_df['combined_text'].tolist()

        # Encode all texts with Sentence-BERT
        aligned_embeddings = {}
        for domain, texts in domain_texts.items():
            aligned_embeddings[domain] = self.sentence_bert_model.encode(texts)

        return aligned_embeddings

    def cross_domain_recommendations(self, source_domain: str, source_item_id: str, target_domain: str, n_recommendations: int = 5):
        """Recommend items across domains based on aligned embeddings"""
        # Get aligned embeddings
        aligned_embeddings = self.multimodal_alignment()

        # Find source item embedding
        if source_domain == 'movies':
            source_idx = self.movies_df[self.movies_df['movie_id'] == source_item_id].index[0]
        elif source_domain == 'books':
            source_idx = self.books_df[self.books_df['book_id'] == source_item_id].index[0]
        elif source_domain == 'food':
            source_idx = self.food_df[self.food_df['recipe_id'] == source_item_id].index[0]
        elif source_domain == 'music':
            source_idx = self.music_df[self.music_df['track_id'] == source_item_id].index[0]
        elif source_domain == 'tv_shows':
            source_idx = self.tv_shows_df[self.tv_shows_df['show_id'] == source_item_id].index[0]
        else:
            raise ValueError(f"Unknown source domain: {source_domain}")

        source_embedding = aligned_embeddings[source_domain][source_idx].reshape(1, -1)
        target_embeddings = aligned_embeddings[target_domain]

        # Calculate similarities
        similarities = cosine_similarity(source_embedding, target_embeddings).flatten()

        # Get top N recommendations
        target_df = getattr(self, f"{target_domain}_df")
        top_indices = similarities.argsort()[-n_recommendations:][::-1]

        return target_df.iloc[top_indices]

# Example usage
if __name__ == "__main__":
    # Load your datasets (assuming they're already generated)
    # For demonstration, we'll use the generated dataframes from your script
    # Option 2: Load from CSV files (if you saved them to Google Drive)
    drive_path = '/content/drive/MyDrive/Recommendation_Datasets/'
    movies_df = pd.read_csv(f'{drive_path}/movies.csv')
    books_df = pd.read_csv(f'{drive_path}/books.csv')
    food_df = pd.read_csv(f'{drive_path}/food.csv')
    music_df = pd.read_csv(f'{drive_path}/music.csv')
    tv_shows_df = pd.read_csv(f'{drive_path}/tv_shows.csv')

    # Initialize the recommender system
    recommender = CrossDomainRecommender(movies_df, books_df, food_df, music_df, tv_shows_df)

    # Prepare data
    recommender.prepare_domain_data()

    # Train TF-IDF models
    recommender.train_tfidf_models()

    # Example: Direct domain query (content-based)
    print("=== Direct Domain Query: Action Movies ===")
    action_movies = recommender.content_based_recommendations('movies', 'action thriller exciting', 3)
    print(action_movies[['title', 'genre', 'rating']])

    # Example: Sentence-BERT recommendations
    print("\n=== Sentence-BERT Recommendations: Romantic Books ===")
    romantic_books = recommender.sentence_bert_recommendations('books', 'romantic love story', 3)
    print(romantic_books[['title', 'author', 'average_rating']])

    # Build cross-domain graph
    recommender.build_cross_domain_graph()

    # Example: Cross-domain recommendations using graph
    print("\n=== Cross-Domain Graph Recommendations ===")
    # Let's say we have a movie we like and want recommendations across domains
    seed_movie = recommender.movies_df.iloc[0]['movie_id']
    graph_recs = recommender.graph_based_recommendations([f"movie_{seed_movie}"], 5)
    print("Graph-based recommendations:", graph_recs)

    # Example: Cross-domain recommendations using multimodal alignment
    print("\n=== Cross-Domain Multimodal Recommendations ===")
    # Get a movie and find similar music
    sample_movie_id = recommender.movies_df.iloc[10]['movie_id']
    cross_domain_recs = recommender.cross_domain_recommendations('movies', sample_movie_id, 'music', 3)
    print("Movie:", recommender.movies_df[recommender.movies_df['movie_id'] == sample_movie_id]['title'].values[0])
    print("Similar music:")
    print(cross_domain_recs[['title', 'artist', 'genre']])

    # Setup RAG system for conversational queries
    recommender.setup_rag_system()

    # Example: Conversational query
    print("\n=== Conversational Query ===")
    conversation_result = recommender.conversational_query("I'm feeling sad and it's raining outside, what should I watch?")
    print("RAG response:", conversation_result)

    # Generate embeddings for movies domain first (ADD THIS LINE)
    _ = recommender.sentence_bert_recommendations('movies', '', 1)

    # Train ranking models (example for movies)
    print("\n=== Training Ranking Model for Movies ===")
    ranking_model, X_test, y_test = recommender.train_ranking_model('movies', 'rating')
    print("Ranking model trained. Test data shape:", X_test.shape)
```

```python
def test_recommendation_prompts(recommender, prompts):
    """Test all recommendation prompts and display results"""

    # Enhanced domain mapping with better keyword matching
    domain_keywords = {
        'movies': ['movie', 'film', 'oscar', 'cinema', 'watch', 'thriller', 'romance', 'comedy', 'drama',
                  'animated', 'holiday', 'courtroom', 'family', 'sports', 'sci-fi', 'tearjerker', 'classic',
                  'time-travel', 'bollywood', 'realistic', 'iconic', 'cinephile', 'rom-com', 'notting hill',
                  'inception', 'dark knight', 'black-and-white'],
        'tv_shows': ['tv show', 'series', 'sitcom', 'k-drama', 'episode', 'season', 'binge', 'netflix', 'hulu',
                    'hbo', 'streaming', 'mini-series', 'reality show', 'detective', 'medical drama', 'gilmore girls',
                    'friends', 'game of thrones', 'breaking bad', 'sherlock'],
        'music': ['music', 'song', 'track', 'album', 'jazz', 'rock', 'pop', 'lo-fi', 'lyrics', 'acoustic',
                 'indie', 'classical', 'electronic', 'soundtrack', 'k-pop', 'meditation', 'piano', 'duet',
                 'taylor swift', 'bts'],
        'books': ['book', 'novel', 'read', 'fantasy', 'romance', 'historical', 'self-improvement', 'thriller',
                 'biography', 'dystopian', 'short story', 'ya novel', 'classic', 'horror', 'philosophical',
                 'gone girl', 'harry potter', 'hunger games'],
        'food': ['recipe', 'food', 'dish', 'vegetarian', 'breakfast', 'soup', 'street food', 'dessert', 'snack',
                'baking', 'meal-prep', 'italian', 'drink', 'low-carb', 'one-pot', 'indian', 'spicy', 'salad',
                'fusion', 'comfort food', 'michelin', 'home alone']
    }

    # Track already recommended items to avoid duplicates
    recommended_items = {
        'movies': set(),
        'tv_shows': set(),
        'music': set(),
        'books': set(),
        'food': set()
    }

    results = {}

    for i, prompt in enumerate(prompts, 1):
        print(f"\n=== Testing Prompt {i}: {prompt} ===")

        # Determine domain based on keywords
        domain = None
        prompt_lower = prompt.lower()

        for dom, keywords in domain_keywords.items():
            if any(keyword in prompt_lower for keyword in keywords):
                domain = dom
                break

        if not domain:
            # If no domain detected, use RAG
            print("Using RAG for cross-domain query")
            try:
                result = recommender.conversational_query(prompt)
                results[i] = {"prompt": prompt, "method": "RAG", "result": result}
                print(f"RAG Result: {result}")
            except Exception as e:
                print(f"Error with RAG: {e}")
                results[i] = {"prompt": prompt, "method": "RAG", "error": str(e)}
            continue

        print(f"Detected domain: {domain}")

        # Use Sentence-BERT for this domain
        try:
            # First ensure embeddings are generated for this domain
            if domain not in recommender.domain_embeddings:
                _ = recommender.sentence_bert_recommendations(domain, "", 1)

            # Get recommendations
            recs = recommender.sentence_bert_recommendations(domain, prompt, 10)  # Get more to filter

            # Filter out already recommended items
            if domain == 'movies' or domain == 'tv_shows':
                title_col = 'title'
            elif domain == 'music':
                title_col = 'title'
            elif domain == 'books':
                title_col = 'title'
            elif domain == 'food':
                title_col = 'name'

            # Filter out duplicates
            unique_recs = []
            for idx, row in recs.iterrows():
                item_id = row[title_col]
                if item_id not in recommended_items[domain]:
                    unique_recs.append(row)
                    recommended_items[domain].add(item_id)
                if len(unique_recs) >= 5:  # We want 5 unique recommendations
                    break

            # If we don't have enough unique recommendations, get more
            if len(unique_recs) < 5:
                # Get additional recommendations
                additional_recs = recommender.sentence_bert_recommendations(domain, prompt, 10)
                for idx, row in additional_recs.iterrows():
                    item_id = row[title_col]
                    if item_id not in recommended_items[domain]:
                        unique_recs.append(row)
                        recommended_items[domain].add(item_id)
                    if len(unique_recs) >= 5:
                        break

            # Convert back to DataFrame
            recs = pd.DataFrame(unique_recs)

            # Format results
            if domain == 'movies':
                result_df = recs[['title', 'genre', 'rating', 'mood']].head(5)
            elif domain == 'tv_shows':
                result_df = recs[['title', 'genre', 'rating', 'mood']].head(5)
            elif domain == 'music':
                result_df = recs[['title', 'artist', 'genre', 'mood']].head(5)
            elif domain == 'books':
                result_df = recs[['title', 'author', 'genre', 'mood']].head(5)
            elif domain == 'food':
                result_df = recs[['name', 'cuisine_type', 'rating', 'mood']].head(5)

            results[i] = {
                "prompt": prompt,
                "domain": domain,
                "method": "Sentence-BERT",
                "result": result_df.to_dict('records')
            }

            print(f"Recommendations:\n{result_df}")

        except Exception as e:
            print(f"Error with Sentence-BERT: {e}")
            # Fall back to TF-IDF
            try:
                recs = recommender.content_based_recommendations(domain, prompt, 10)

                # Filter out duplicates
                unique_recs = []
                for idx, row in recs.iterrows():
                    item_id = row[title_col]
                    if item_id not in recommended_items[domain]:
                        unique_recs.append(row)
                        recommended_items[domain].add(item_id)
                    if len(unique_recs) >= 5:
                        break

                # Convert back to DataFrame
                recs = pd.DataFrame(unique_recs)

                if domain == 'movies':
                    result_df = recs[['title', 'genre', 'rating', 'mood']].head(5)
                elif domain == 'tv_shows':
                    result_df = recs[['title', 'genre', 'rating', 'mood']].head(5)
                elif domain == 'music':
                    result_df = recs[['title', 'artist', 'genre', 'mood']].head(5)
                elif domain == 'books':
                    result_df = recs[['title', 'author', 'genre', 'mood']].head(5)
                elif domain == 'food':
                    result_df = recs[['name', 'cuisine_type', 'rating', 'mood']].head(5)

                results[i] = {
                    "prompt": prompt,
                    "domain": domain,
                    "method": "TF-IDF",
                    "result": result_df.to_dict('records')
                }

                print(f"Fallback to TF-IDF Recommendations:\n{result_df}")

            except Exception as e2:
                print(f"Error with TF-IDF: {e2}")
                # Final fallback to RAG
                try:
                    result = recommender.conversational_query(prompt)
                    results[i] = {"prompt": prompt, "method": "RAG", "result": result}
                    print(f"Final fallback to RAG: {result}")
                except Exception as e3:
                    print(f"Error with RAG: {e3}")
                    results[i] = {"prompt": prompt, "method": "RAG", "error": str(e3)}

    return results

# Improved RAG system with better filtering
def setup_rag_system(self):
    """Set up the RAG system for conversational queries with domain filtering"""
    # Create separate vector stores for each domain
    self.domain_vector_stores = {}

    # Movies
    movie_data = []
    for idx, row in self.movies_df.iterrows():
        movie_data.append(f"Movie: {row['title']}. Description: {row['description']}. Genre: {row['genre']}. Mood: {row['mood']}")
    self.domain_vector_stores['movies'] = FAISS.from_texts(movie_data, self.rag_embeddings)

    # TV Shows
    tv_data = []
    for idx, row in self.tv_shows_df.iterrows():
        tv_data.append(f"TV Show: {row['title']}. Description: {row['description']}. Genre: {row['genre']}. Mood: {row['mood']}")
    self.domain_vector_stores['tv_shows'] = FAISS.from_texts(tv_data, self.rag_embeddings)

    # Music
    music_data = []
    for idx, row in self.music_df.iterrows():
        music_data.append(f"Music: {row['title']}. Artist: {row['artist']}. Genre: {row['genre']}. Mood: {row['mood']}")
    self.domain_vector_stores['music'] = FAISS.from_texts(music_data, self.rag_embeddings)

    # Books
    book_data = []
    for idx, row in self.books_df.iterrows():
        book_data.append(f"Book: {row['title']}. Author: {row['author']}. Genre: {row['genre']}. Mood: {row['mood']}")
    self.domain_vector_stores['books'] = FAISS.from_texts(book_data, self.rag_embeddings)

    # Food
    food_data = []
    for idx, row in self.food_df.iterrows():
        food_data.append(f"Food: {row['name']}. Cuisine: {row['cuisine_type']}. Category: {row['category']}. Mood: {row['mood']}")
    self.domain_vector_stores['food'] = FAISS.from_texts(food_data, self.rag_embeddings)

    # Create a general retriever for cross-domain queries
    all_data = movie_data + tv_data + music_data + book_data + food_data
    self.general_vector_store = FAISS.from_texts(all_data, self.rag_embeddings)

    # Create retrievers
    self.domain_retrievers = {domain: store.as_retriever(search_kwargs={"k": 3}) for domain, store in self.domain_vector_stores.items()}
    self.general_retriever = self.general_vector_store.as_retriever(search_kwargs={"k": 3})

def conversational_query(self, query: str):
    """Handle conversational queries using RAG with domain detection"""
    query_lower = query.lower()

    # Try to detect domain
    domain = None
    if any(keyword in query_lower for keyword in ['movie', 'film', 'cinema']):
        domain = 'movies'
    elif any(keyword in query_lower for keyword in ['tv', 'show', 'series', 'episode']):
        domain = 'tv_shows'
    elif any(keyword in query_lower for keyword in ['music', 'song', 'album', 'track']):
        domain = 'music'
    elif any(keyword in query_lower for keyword in ['book', 'novel', 'read']):
        domain = 'books'
    elif any(keyword in query_lower for keyword in ['food', 'recipe', 'dish', 'cook']):
        domain = 'food'

    # Use domain-specific retriever if detected
    if domain and domain in self.domain_retrievers:
        docs = self.domain_retrievers[domain].get_relevant_documents(query)
    else:
        # Use general retriever for cross-domain queries
        docs = self.general_retriever.get_relevant_documents(query)

    # Format the response
    response = "Based on your query, here are some recommendations:\n\n"
    for i, doc in enumerate(docs):
        # Extract the most relevant part of the document
        content = doc.page_content
        if "Movie:" in content:
            title = content.split("Movie: ")[1].split(". ")[0]
            genre = content.split("Genre: ")[1].split(". ")[0] if "Genre: " in content else "Unknown"
            response += f"{i+1}. Movie: {title} ({genre})\n"
        elif "TV Show:" in content:
            title = content.split("TV Show: ")[1].split(". ")[0]
            genre = content.split("Genre: ")[1].split(". ")[0] if "Genre: " in content else "Unknown"
            response += f"{i+1}. TV Show: {title} ({genre})\n"
        elif "Music:" in content:
            title = content.split("Music: ")[1].split(". ")[0]
            artist = content.split("Artist: ")[1].split(". ")[0] if "Artist: " in content else "Unknown"
            response += f"{i+1}. Music: {title} by {artist}\n"
        elif "Book:" in content:
            title = content.split("Book: ")[1].split(". ")[0]
            author = content.split("Author: ")[1].split(". ")[0] if "Author: " in content else "Unknown"
            response += f"{i+1}. Book: {title} by {author}\n"
        elif "Food:" in content:
            name = content.split("Food: ")[1].split(". ")[0]
            cuisine = content.split("Cuisine: ")[1].split(". ")[0] if "Cuisine: " in content else "Unknown"
            response += f"{i+1}. Food: {name} ({cuisine})\n"
        else:
            response += f"{i+1}. {content[:200]}...\n"

    return response

# Add these methods to the class
CrossDomainRecommender.setup_rag_system = setup_rag_system
CrossDomainRecommender.conversational_query = conversational_query

# Now run the test with the improved system
print("Setting up improved RAG system...")
recommender.setup_rag_system()

print("Testing prompts with improved system...")
test_results = test_recommendation_prompts(recommender, all_prompts)

# Save results
import json
with open('/content/drive/MyDrive/Recommendation_Datasets/test_results_improved.json', 'w') as f:
    json.dump(test_results, f, indent=2)

print("Testing completed! Results saved to test_results_improved.json")
```
